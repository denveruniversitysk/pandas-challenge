```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Creating variable to define the length of a string to calculate total number of schools
count_schools = len(school_data_complete['School ID'].unique())

# Creating variable to define the length of a string to calculate total number of students
count_students = len(school_data_complete['Student ID'].unique())

# Calculating total budget
total_budget = school_data['budget'].sum()

# Calculating average math score
average_math = student_data['math_score']. mean()

# Calculating average reading score
average_reading = student_data['reading_score'].mean()

# Calculating percentage for passing reading score
passing_reading_score = student_data.loc[student_data['reading_score'] >= 70]['reading_score'].count()
reading_percent = passing_reading_score/count_students

# Calculating percentage for passing math score
passing_math_score = student_data.loc[student_data['math_score'] >=70]['math_score'].count()
math_percent = passing_math_score/count_students

# Calculating percentage for students passing math and reading
percent_overall = student_data[(student_data['math_score'] >=70) & (student_data['reading_score'] >=70)]['student_name'].count()/count_students

# Creating DataFrame
dist_summary = pd.DataFrame({
    
    "Total Schools": [count_schools],
    "Total Students": [count_students],
    "Total Budget": [total_budget],
    "Average Math Score": [average_math],
    "Average Reading Score": [average_reading],
    "% Passing Math": [math_percent],
    "% Passing Reading": [reading_percent],
    "% Overall Passing": [percent_overall]})

dist_summary

# Formatting DataFrame
dist_summary.style.format({"Total Budget":"${:,.2f}",
                          "Average Reading Score":"{:.5f}",
                          "Average Math Score":"{:.6f}",
                          "% Passing Math":"{:,.6%}",
                          "% Passing Reading": "{:.6%}", 
                           "% Overall Passing": "{:.6%}"})
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Splitting data into groups by school name
name_school = school_data_complete.set_index('school_name').groupby(['school_name'])

# Defining school type
school_type = school_data.set_index('school_name')['type']

# Calculating total students
student_total = name_school["Student ID"].count()

# Calculating total school budget
total_budget = school_data.set_index('school_name')['budget']

# Calculating per student budget
budget_per_stu = total_budget/student_total

# Calculating average math score
average_math_score = name_school["math_score"].mean()

# Calculating average reading score
average_reading_score = name_school["reading_score"].mean()

# Calculating percent of students passing math
percent_math = school_data_complete.loc[school_data_complete["math_score"] >=70].groupby('school_name')['Student ID'].count()/student_total

# Calculating percent of students passing reading
percent_reading = school_data_complete.loc[school_data_complete["reading_score"] >=70].groupby('school_name')['Student ID'].count()/student_total

# Calculating percent of students passing math and reading
overall_pass_percent = school_data_complete.loc[school_data_complete["math_score"] >=70].loc[school_data_complete["reading_score"] >=70].groupby('school_name')['Student ID'].count()/student_total

# Creating DataFrame
sch_summary = pd.DataFrame({
    "School Type": school_type,
    "Total Students": student_total,
    "Total School Budget": total_budget,
    "Per Student Budget": budget_per_stu,
    "Average Math Score": average_math_score,
    "Average Reading Score": average_reading_score,
    "% Passing Math": percent_math,
    "% Passing Reading": percent_reading,
    "% Overall Passing": overall_pass_percent
})

sch_summary

# Formatting DataFrame
sch_summary.style.format({"Total School Budget":"${:,}",
                          "Per Student Budget":"${:.2f}",
                          "Average Reading Score":"{:.6f}",
                          "Average Math Score":"{:.6f}",
                          "% Passing Math":"{:,.6%}",
                          "% Passing Reading": "{:.6%}", 
                           "% Overall Passing": "{:.6%}"})
```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
# Sorting values to calculate key metrics for top performing schools
sch_summ_sort = sch_summary.sort_values("% Overall Passing", ascending=False)[:5]
sch_summ_sort

# Formatting table
sch_summ_sort.head().style.format({"Total School Budget": "${:,.2f}", 
                                   "Per Student Budget": "${:.2f}", 
                                   "Average Math Score":"{:.6f}",
                                   "Average Reading Score":"{:.6f}",
                                   "% Passing Math": "{:.6%}", 
                                   "% Passing Reading": "{:.6%}",
                                   "% Overall Passing": "{:.6%}"})
```

## Bottom Performing Schools (By % Overall Passing)

* Sort and display the five worst-performing schools by % overall passing.

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Calculating mean math score for 9th grade students
math_ninth = student_data.loc[student_data['grade'] == '9th'].groupby('school_name')["math_score"].mean()

# Calculating mean math score for 10th grade students
math_tenth = student_data.loc[student_data['grade'] == '10th'].groupby('school_name')["math_score"].mean()

# Calculating mean math score for 11th grade students
math_eleventh = student_data.loc[student_data['grade'] == '11th'].groupby('school_name')["math_score"].mean()

# Calculating mean math score for 12th grade students
math_twelfth = student_data.loc[student_data['grade'] == '12th'].groupby('school_name')["math_score"].mean()

# Building DataFrame
math_sc_by_grade = pd.DataFrame({
    "9th": math_ninth,
    "10th": math_tenth,
    "11th": math_eleventh,
    "12th": math_twelfth
})
math_sc_by_grade = math_sc_by_grade[['9th','10th', '11th', '12th']]

# Setting index for DataFrame
math_sc_by_grade.index.name = None

math_sc_by_grade
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
# Calculating mean reading score for 9th grade students
reading_ninth = student_data.loc[student_data['grade'] == '9th'].groupby('school_name')["reading_score"].mean()

# Calculating mean reading score for 10th grade students
reading_tenth = student_data.loc[student_data['grade'] == '10th'].groupby('school_name')["reading_score"].mean()

# Calculating mean reading score for 11th grade students
reading_eleventh = student_data.loc[student_data['grade'] == '11th'].groupby('school_name')["reading_score"].mean()

# Calculating mean reading score for 12th grade students
reading_twelfth = student_data.loc[student_data['grade'] == '12th'].groupby('school_name')["reading_score"].mean()

# Building DataFrame
reading_sc_by_grade = pd.DataFrame({
    "9th": reading_ninth,
    "10th": reading_tenth,
    "11th": reading_eleventh,
    "12th": reading_twelfth
})
reading_sc_by_grade = reading_sc_by_grade[['9th','10th', '11th', '12th']]

# Setting index for DataFrame
reading_sc_by_grade.index.name = None

reading_sc_by_grade
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Creating bins and defining labels
bins = [0, 585, 630, 645, 675]
labels = ["<$584","$585-629","$630-644","$645-675"]

# Defining variable
spending_by_student = sch_summary
```

```python
# Using pd.cut method to sort bin values into discrete variables
spending_by_student["Spending Ranges (Per Student)"] = pd.cut(budget_per_stu,bins,labels=labels)
```

```python
# Calculating average math score by school spending
spending_math = spending_by_student.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]

# Calculating average reading score by school spending
spending_reading = spending_by_student.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]

# Calculating percent of students passing math based on school spending
percent_pass_math = spending_by_student.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]

# Calculating percent of students passing reading based on school spending
percent_pass_read = spending_by_student.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]

# Calculating percent of students passing math and reading based on school spending
overall_pass_percent = spending_by_student.groupby(["Spending Ranges (Per Student)"]).mean()["% Overall Passing"]
```

```python
# Building DataFrame
scores_per_spend = pd.DataFrame({
    "Average Math Score": spending_math,
    "Average Reading Score": spending_reading,
    "% Passing Math": percent_pass_math,
    "% Passing Reading": percent_pass_read,
    "% Overall Passing": overall_pass_percent
})
scores_per_spend

# Formatting DataFrame
scores_per_spend.style.format({"Average Math Score":"{:.2f}",
                              "Average Reading Score":"{:.2f}",
                              "% Passing Math":"{:.2%}",
                              "% Passing Reading":"{:.2%}",
                              "% Overall Passing":"{:.2%}"})
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# Creating bins and defining labels
bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)","Medium (1000-2000)","Large (2000-5000)"]

# Defining variable
scores_by_size = sch_summary

# Using pd.cut method to sort bin values into discrete variables
scores_by_size["School Size"] = pd.cut(student_total,bins,labels=labels)

# Calculating math scores by school size
score_math = scores_by_size.groupby(["School Size"]).mean()["Average Math Score"]

# Calculating reading scores by school size
score_reading = scores_by_size.groupby(["School Size"]).mean()["Average Reading Score"]

# Calculating percent of students passing math by school size
score_pass_math = scores_by_size.groupby(["School Size"]).mean()["% Passing Math"]

# Calculating percent of students passing reading by school size
score_pass_read = scores_by_size.groupby(["School Size"]).mean()["% Passing Reading"]

# Calculating percent of students passing math and reading by school size
overall_percent = scores_by_size.groupby(["School Size"]).mean()["% Overall Passing"]
```

```python
# Building DataFrame
scores_per_size = pd.DataFrame({
    "Average Math Score": score_math,
    "Average Reading Score": score_reading,
    "% Passing Math": score_pass_math,
    "% Passing Reading": score_pass_read,
    "% Overall Passing": overall_percent
})
scores_per_size

# Formatting DataFrame
scores_per_size.style.format({"Average Math Score":"{:.6f}",
                              "Average Reading Score":"{:.6f}",
                              "% Passing Math":"{:.6%}",
                              "% Passing Reading":"{:.6%}",
                              "% Overall Passing":"{:.6%}"})
```
